#### Read & Define Dataset

```python
import pandas as pd

train_df = pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='latin1')
test_df = pd.read_csv('testdata.manual.2009.06.14.csv', encoding='latin1')
```

```python
train_df = train_df.sample(frac=1).reset_index(drop=True)
train_df
```

```python
test_df = test_df.sample(frac=1).reset_index(drop=True)
test_df
```

```python
train_df.columns = ['polarity', 'id', 'datetime', 'query', 'username', 'tweet']
train_df = train_df.reset_index(drop=True)
train_df
```

```python
test_df.columns = ['polarity', 'id', 'datetime', 'query', 'username', 'tweet']
test_df = test_df.reset_index(drop=True)
test_df
```

```python
train_df = train_df[['polarity', 'tweet']]
train_df
```

```python
test_df = test_df[['polarity', 'tweet']]
test_df
```

#### Preprocess

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
import string
import re

lemmatizer = WordNetLemmatizer()
stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def preprocess_text(text):
    global lemmatizer , stemmer , stop_words
    
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)
    text = re.sub(r'\d+', '', text)
    text = re.sub(r'[@#]', '', text)

    tokens = nltk.word_tokenize(text.lower())
    tokens = [token for token in tokens if token not in stop_words]
    tokens = [token for token in tokens if token not in string.punctuation]
    tokens = [lemmatizer.lemmatize(token) for token in tokens]
    tokens = [stemmer.stem(token) for token in tokens]
    
    text = ' '.join(tokens)
    
    return text
```

```python
train_df['tweet'] = train_df['tweet'].apply(preprocess_text)
```

#### Define X , y Train & Test

```python
X_train = train_df['tweet']
y_train = train_df['polarity']

new_test_df = test_df[test_df['polarity'] != 2]
X_test = new_test_df['tweet']
y_test = new_test_df['polarity']
```

#### Count Vectorizer

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()

X_train = vectorizer.fit_transform(X_train)

X_test = vectorizer.transform(X_test)
```

#### Train Model

```python
from sklearn.naive_bayes import MultinomialNB

model = MultinomialNB()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
y_pred
```

#### Confusion Matrix

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns

cm = confusion_matrix(y_test, y_pred)

plt.figure()
sns.heatmap(cm, annot=True, fmt="d", cmap="Greens", xticklabels=['Negative','Positive'], yticklabels=['Negative','Positive'])
            
plt.xlabel("Predicted")
plt.ylabel("Real")
plt.title("Confusion Matrix")
plt.show()
```

#### Results

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score

accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred, average="macro")
recall = recall_score(y_test, y_pred, average="macro")

print('Accuracy: ',accuracy)
print('Precision: ',precision)
print('Recall: ',recall)
```
